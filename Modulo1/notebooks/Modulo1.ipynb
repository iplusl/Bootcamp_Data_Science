## Desafio: reposicionar a legenda. Dentro? Fora? Onde?

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = usaveis.T.plot(figsize=(14,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor Gasto em Saúde(R$) x Tempo")
plt.legend(loc='best', bbox_to_anchor=(1, 1), ncol=3)
plt.show()
```

## Desafio: Selecionar e plotar 5 estados

Selecionando na mão os 5 estados:

```python
ax = usaveis.T[["35 São Paulo", "33 Rio de Janeiro",
                "53 Distrito Federal", "31 Minas Gerais", "41 Paraná"]].plot(figsize=(14,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor Gasto em Saúde(R$) x Tempo")
plt.legend(loc='best', bbox_to_anchor=(1, 1), ncol=1)
plt.show()
```

Selecionando aleatoriamente os 5 estados:

```python
import random

estados = usaveis.T.columns.to_list()
estados_aleatorios = random.sample(estados, 5)

ax = usaveis.T[estados_aleatorios].plot(figsize=(14,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor Gasto em Saúde(R$) x Tempo")
plt.legend(loc='best', bbox_to_anchor=(1, 1), ncol=1)
plt.show()
```

Selecionando os 5 estados que tiveram maior média de gasto:

```python
estados_media = usaveis.T.mean().sort_values(ascending = False)
estados_media.index[0:5].to_list()

ax = usaveis.T[estados_media.index[0:5].to_list()].plot(figsize=(14,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Top 5 Estados em Valor Gasto em Saúde(R$) x Tempo")
plt.legend(loc='best', bbox_to_anchor=(1, 1), ncol=1)
plt.show()
```

## Desafio: Mudar a palete de cores com o matplotlib

Para mudar as cores podemos usar o plt.style.use() e escolher um tema disponível na [documentação](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html).

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
plt.style.use('seaborn-darkgrid')

ax = ordenado_por_total.head(5).T.plot(figsize=(14,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Top 5 Estados em Valor Gasto com Saúde(R$) x Tempo")
plt.legend(loc='upper left', ncol=1)
plt.show()
```

```python
plt.style.use('default')
```

Mas também podemos escolher as cores manualmente como mostro no código abaixo:

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = ordenado_por_total.head(5).T.plot(figsize=(14,8),
                                       color=["green", "red", "blue", "gray", "yellow"])
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Top 5 Estados em Valor Gasto com Saúde(R$) x Tempo")
plt.legend(loc='upper left', ncol=1)
plt.show()
```

No caso acima eu coloquei manualmente o nome das cores e podemos acessar a lista na [documentação](https://matplotlib.org/stable/gallery/color/named_colors.html).<br> Porém também é possível definir as cores com uma string hex ('#008000'), tupla RGB ou RGBA ((0,1,0,1)) ou ainda com uma string de intensidade da escala de cinza ('0.8').

## Desafio: Adiciona uma coluna mostrando a região de cada estado

```python
regioes = []

for estado in ordenado_por_total.index.to_list():
    if estado[0] == '1':
        regioes.append("Norte")
    elif estado[0] == '2':
        regioes.append("Nordeste")
    elif estado[0] == '3':
        regioes.append("Sudeste")
    elif estado[0] == '4':
        regioes.append("Sul")
    elif estado[0] == '5':
        regioes.append("Centro-Oeste")

ordenado_por_total["Região"] = regioes
ordenado_por_total.head()
```

## Desafio: Deixar o gráfico mais bonito. A mensagem mais clara.
ordenado_por_total.head(5).T.plot(figsize = (10,6))

```python
plt.style.use('seaborn')

colunas_sem_regiao = ordenado_por_total.columns[:-1].to_list()

ax = ordenado_por_total[colunas_sem_regiao].head(5).T.plot(figsize = (10,6))

ticks_loc = ax.get_yticks().tolist()
ax.yaxis.set_major_locator(ticker.FixedLocator(ticks_loc))
ax.set_yticklabels(['{:,.0f}'.format(x) + 'Milhões' for x in ax.get_yticks()/1000000])
ax.tick_params(axis = 'both', which = 'major', labelsize = 12)

plt.title("Top 5 Estados em Valor Gasto com Saúde(R$) x Tempo", fontdict = {"fontsize":16, "fontweight":"bold"})
plt.legend(loc='upper left', fontsize=12)
plt.show()
```
