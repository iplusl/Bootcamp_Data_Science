## Desafio: Adiciona uma coluna mostrando a região de cada estado

```python
meses_filtrados = ordenado_por_total.columns[47:150]
ordenado_por_total[meses_filtrados].head(3).T.plot(figsize = (10,6))
```

No gráfico acima temos a impressão que os dados para o estado do Paraná entre 2012/Jan e 2013/Set estão em torno de 50 milhões. Mas essa informação pode estar sendo tendenciosa e por isso devemos nos atentar aos limites dos eixos. Nesse caso podemos observar que no eixo y não temos o valor de 0, e por isso não podemos ter certeza dos valores abaixo de 100 milhões.<br>

Vamos "forçar" o pyplot a usar os limites de y que desejamos com a função plt.ylim()

```python
ordenado_por_total[meses_filtrados].head(3).T.plot(figsize = (10,6))
plt.ylim(0, 350)
```

Agora que temos o 0 do eixo y, fica mais claro ainda a desinformação presente no gráfico anterior. Antes tinhamos percebido que os valores para o estado do Paraná entre 2012/Jan e 2013/Set estavam em torno de 50 milhões. Nesse nosso gráfico vemos claramente que não, na verdade os valores estão em torno de 75 milhões.

```python
ordenado_por_total[meses_filtrados].head(3).T.plot(figsize = (10,6))
plt.ylim(100, 350)
```

No gráfico acima colocamos o limite inferior do eixo y em 100 milhões, o que deixou os dados mais tendenciosos ainda. Algo que poderia levar um leitor desatento a crer que antes de meados de 2016 o estado do Paraná não tinha gastos com saúde.

```python
ordenado_por_total[ordenado_por_total.columns[:]].head(3).T.plot(figsize = (10,6))
plt.ylim(0, 350)
```

Esse cuidado não deve se limitar apenas aos eixos, mas também devemos ficar atentos ao tamanho do gráfico em si. No gráfico acima podemos ver claramente uma tendência crecente nos três estados.

```python
ordenado_por_total.head(3).T.plot(figsize = (6,6))
```
